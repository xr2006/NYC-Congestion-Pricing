### 1. Retrieve CV and coefficients

```python
data = pd.read_csv(root + '4.Distributional Impacts/(1226) shapley_decomposition_CS.csv')
prediction = pd.read_csv(root+'3.Model Calibration/(1226) trip_prediction.csv')
prediction['origin_county'] = prediction['origin_county'].map(lambda x:'0'+str(x) if len(str(x))==4 else str(x))
choice_data_25 = pd.read_csv('/Users/ryan/Documents/01.Research Projects/18.NYC Congestion Pricing/3.Model Calibration/(1226) choice_dataset_2025.csv')

market_id_colum = ['origin_county','Pop_group','Trip_purpose','Time_period']
transit_num = prediction[prediction['mode']=='transit'].groupby(market_id_colum)['Trip_num'].sum().reset_index()
transit_num.rename(columns={'Trip_num':'Transit_num'},inplace=True)

data = pd.merge(data,transit_num,on=market_id_colum,how='left')
data['Trip_num_transit'] = data['Transit_num'].fillna(0)
# data = data[data['Trip_num_transit']>0]
data['Transit_share'] = data['Trip_num_transit']/data['Trip_num']
```

```python
Segment_lst = ['NotLowIncome_Commute_Peak','NotLowIncome_Commute_Overnight','NotLowIncome_Other_Peak','NotLowIncome_Other_Overnight',
               'LowIncome_Commute_Peak','LowIncome_Commute_Overnight','LowIncome_Other_Peak','LowIncome_Other_Overnight',
               'Senior_Commute_Peak','Senior_Commute_Overnight','Senior_Other_Peak','Senior_Other_Overnight',
               'Student_Commute_Peak','Student_Commute_Overnight','Student_Other_Peak','Student_Other_Overnight']

segment_name = Segment_lst[10]
file_path = root + '2.Joint mode and destination modeling/Model Results 1226/IPDL_%s.xlsx'%segment_name
df = pd.read_excel(file_path, sheet_name="Coefficients")
attr_name = df["term"].tolist()
beta_name_R = [x for x in attr_name if x[:6]!='factor']
attr_name = [s.replace('log_s_s.county.','log_s_s(county)') for s in attr_name]
attr_name = [s.replace('log_s_s.mode.','log_s_s(mode)') for s in attr_name]
attr_name = [s.replace('County_36061.0','County_36061-0') for s in attr_name]
attr_name = [s.replace('County_36061.1','County_36061-1') for s in attr_name]
beta_name_Python = [x for x in attr_name if x[:6]!='factor']

coeff_dict = {}
for segment_name in Segment_lst:
    file_path = root + '2.Joint mode and destination modeling/Model Results 1226/IPDL_%s.xlsx'%segment_name
    df = pd.read_excel(file_path, sheet_name="Coefficients")    
    coeff = df[df['term'].isin(beta_name_R)]['estimate'].values
    population,purpose,period = segment_name.split('_')
    coeff_dict[(population,purpose,period)] = coeff
```

```python
def get_coeff(record):
    coeff = coeff_dict[(record['Pop_group'],record['Trip_purpose'],record['Time_period'])]
    cost_new = coeff[beta_name_Python.index('cost_new')]
    cost_new_NYC = coeff[beta_name_Python.index('cost_new_NYC')]
    transit_wt = coeff[beta_name_Python.index('transit_wt')]
    transit_wt_NYC = coeff[beta_name_Python.index('transit_wt_NYC')]
    return pd.Series([cost_new,cost_new_NYC,transit_wt,transit_wt_NYC])

data[['cost_new','cost_new_NYC','transit_wt','transit_wt_NYC']] = data.apply(get_coeff,axis=1)
data['beta_cost'] = data['cost_new'] + data['cost_new_NYC']*data['origin_region'].map(lambda x:1 if x in ['CRZ','Manhattan','NYC'] else 0)
data['beta_wt'] = data['transit_wt'] + data['transit_wt_NYC']*data['origin_region'].map(lambda x:1 if x in ['CRZ','Manhattan','NYC'] else 0)
```

```python
data['CV_1min_transit_wt'] = data['beta_wt']/data['beta_cost'] * data['Transit_share'] * data['Trip_num']
data['CV_1unit_transit_cost'] = 1 * data['Transit_share'] * data['Trip_num']

diff = data.copy()
diff['wt_compensation'] = diff['CS_toll']/(diff['CV_1min_transit_wt'])
diff['fare_compensation'] = diff['CS_toll']/(diff['CV_1unit_transit_cost'])
```

```python
print('Total welfare loss:', -diff['CS_toll'].sum()*365/1e6)
print('Total welfare loss in NJS:', -diff[diff['origin_region']=='NJS']['CS_toll'].sum()*365/1e6)
print('Total welfare loss in NYC:', -diff[diff['origin_region'].isin(['CRZ','NYC','Manhattan'])]['CS_toll'].sum()*365/1e6)
print('Total welfare loss in other regions:', -diff[diff['origin_region'].isin(['CTS & PAS','NYS'])]['CS_toll'].sum()*365/1e6)

diff = diff[diff['Trip_num_transit']>0]
```

### Transit Strategy: NYC Part

```python
condition = (diff['origin_region'].isin(['Manhattan','NYC','CRZ'])) & (diff['CS_toll']<0)
change = diff[condition]['CS_toll'].sum()/diff[condition]['CV_1min_transit_wt'].sum()
print(f'Average decrease of transit waiting time in NYC to offset the negative CV: {-change:.2f} min')

aa = choice_data_25[(choice_data_25['mode']=='transit')&(choice_data_25['origin_region'].isin(['Manhattan','NYC','CRZ']))]
avg_wt = ((aa['transit_wt']*aa['Trip_num'])/aa['Trip_num'].sum()).sum()
print(f'Average transit waiting time in NYC: {avg_wt:.2f} min')

aa['cost*trip'] = aa['cost_new']*aa['Trip_num']
bb = aa.groupby('Pop_group').agg({'cost*trip':'sum','Trip_num':'sum'})
avg_cost = bb['cost*trip']/bb['Trip_num']
print('Average transit fare in NYC:',avg_cost)
```

```python
compensation_list = diff[condition].sort_values('wt_compensation')

cols_to_average = ['CS_toll','CV_1unit_transit_cost', 'CV_1min_transit_wt']

# 先计算每列的加权值 (value * weight)
for col in cols_to_average:
    compensation_list[f'{col}_weighted'] = compensation_list[col] * compensation_list['Trip_num']

# 按组汇总
grouped = compensation_list.groupby(['Pop_group', 'origin_county']).agg({
    'Trip_num': 'sum',
    'Trip_num_transit': 'sum',
    **{f'{col}_weighted': 'sum' for col in cols_to_average}
}).reset_index()

# 计算加权平均 (sum of weighted values / sum of weights)
for col in cols_to_average:
    grouped[col] = grouped[f'{col}_weighted'] / grouped['Trip_num']
    grouped.drop(columns=[f'{col}_weighted'], inplace=True)

compensation_list = grouped
```

```python
# Strategy: Compensate by transit waiting time, and then use segment-level fare to compensate the rest.

# Step 1: Set the decrease of transit waiting time
df = pd.DataFrame(columns=["WT_decrease","Discount(Pareto)","Subsidy Per Day","Subsidy Per Year"])

wt_lst = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
for i in wt_lst:
    compensation_list_ = compensation_list.copy()
    # Step 2: Update CV
    CV = compensation_list_['CS_toll'] + i*compensation_list_['CV_1min_transit_wt']
    # Step 3: Remove markets with positive CV
    CV[CV>=0] = 0
    compensation_list_['CS_toll'] = CV
    uncompensated_trips = compensation_list_['Trip_num'].sum()
    uncompensated_transit_trips = compensation_list_['Trip_num_transit'].sum()
    # Step 4: Calculate population segment-level fare discounts to ensure lower bound (pareto)
    compensation_list_['fare_compensation'] = compensation_list_['CS_toll']/compensation_list_['CV_1unit_transit_cost']
    fare_discount = compensation_list_.groupby('Pop_group')['fare_compensation'].min() *-1
    compensation_list_['pop_fare_discount'] = compensation_list_['Pop_group'].map(lambda x:fare_discount.loc[x])
    subsidy = (compensation_list_['pop_fare_discount']*compensation_list_['Trip_num_transit']).sum()
    fare_discount = fare_discount.map(lambda x:round(x,2))
    discount = {'NotLowIncome': 0, 'LowIncome': 0, 'Senior': 0, 'Student': 0}
    discount.update(fare_discount.to_dict())
    # Step 5: Append new row
    df.loc[len(df)] = [i,discount,round(subsidy),round(subsidy*365/1e6,2)]
```

### Predict Transit Ridership

```python
def update_attributes(data,wt_decrease_NYC,discount_NYC,wt_decrease_NJS,discount_NJS):
    choice_data = data.copy()
    mask_transit = choice_data['mode']=='transit'
    mask_NYC = choice_data['origin_region'].isin(['Manhattan','CRZ','NYC'])
    mask_NJS = choice_data['origin_region'].isin(['NJS'])
    choice_data.loc[mask_transit&mask_NYC,'transit_wt_NYC'] -= float(wt_decrease_NYC)
    choice_data.loc[mask_transit&mask_NJS,'transit_wt'] -= float(wt_decrease_NJS)
    for segment in ['NotLowIncome','LowIncome','Senior','Student']:
        mask_seg = choice_data['Pop_group']==segment
        choice_data.loc[mask_transit&mask_NYC&mask_seg,'cost_new_NYC'] -= float(discount_NYC[segment])
        choice_data.loc[mask_transit&mask_NJS&mask_seg,'cost_new'] -= float(discount_NJS[segment])
    choice_data['transit_wt_NYC'] = choice_data['transit_wt_NYC'].map(lambda x: max(0,x))
    choice_data['transit_wt'] = choice_data['transit_wt'].map(lambda x: max(0,x))
    choice_data['cost_new_NYC'] = choice_data['cost_new_NYC'].map(lambda x: max(0,x))
    choice_data['cost_new'] = choice_data['cost_new'].map(lambda x: max(0,x))
    
    return choice_data
```

```python
def prediction_with_coeff_adj(data,Segment_lst,adj={'constant_driving':0,'constant_transit':0,'constant_ondemand':0,'constant_biking':0,'constant_walking':0,'County_36061-1':0}):
    # initialize columns once (optional but recommended)
    choice_data = data.copy()
    pred_cols = ['log_s_s0_pred', 'share_0_shared', 'Trip_num_pred']
    # For loop segments
    for segment_name in Segment_lst:
        pop_group, trip_purpose, time_period = segment_name.split('_')
        # ---- load coefficients ----
        file_path = (
            root
            + f'2.Joint mode and destination modeling/Model Results 1226/'
              f'IPDL_{segment_name}.xlsx'
        )
        df = pd.read_excel(file_path, sheet_name="Coefficients")
        attr_name = df["term"].tolist()
        attr_name = [s.replace('log_s_s.county.', 'log_s_s(county)') for s in attr_name]
        attr_name = [s.replace('log_s_s.mode.', 'log_s_s(mode)') for s in attr_name]
        attr_name = [s.replace('County_36061.0', 'County_36061-0') for s in attr_name]
        attr_name = [s.replace('County_36061.1', 'County_36061-1') for s in attr_name]
        coeff = df['estimate'].values
        market_constant_name = [x for x in attr_name if x.startswith('factor')]
        mask = (
            (choice_data['Pop_group'] == pop_group) &
            (choice_data['Trip_purpose'] == trip_purpose) &
            (choice_data['Time_period'] == time_period)
        )
        piece = choice_data.loc[mask]
        
        # ---- apply adjustment ---
        coeff_full = np.tile(coeff, (len(piece), 1))
        coeff_adj = coeff_full.copy()
        # Mode constant adjustment is only for trips from/to CRZ
        condition = (piece['destination_county']=='36061-1')|(piece['origin_county']=='36061-1')
        for attr in list(adj.keys())[:-1]:
            coeff_adj[condition,attr_name.index(attr)] += adj[attr]
        # County constant adjustment is for all trips
        coeff_adj[:,attr_name.index('County_36061-1')] += adj['County_36061-1']

        # ---- predict ----
        preds = predict_IPDL_trip(
            piece,
            attr_name,
            market_constant_name,
            segment_name,
            coeff_adj,
            print_metrics = False
        )
    
        # preds must be aligned row-wise with `piece`
        # ---- assign back safely ----
        choice_data.loc[mask, pred_cols] = preds.iloc[:,:-1].values
    
    return choice_data

def extract_r_block(r_path,
                    start_marker = 'source("2.Joint mode and destination modeling/(1226) utils.R")',
                    end_marker = 'county_vars <-',):
    
    lines = r_path.read_text(encoding="utf-8").splitlines(True)
    start_idx = None
    end_idx = None

    for i, line in enumerate(lines):
        if start_idx is None and start_marker in line:
            start_idx = i
            continue
        if start_idx is not None and end_marker in line:
            end_idx = i
            break

    if start_idx is None:
        raise ValueError("Start marker not found.")
    if end_idx is None:
        raise ValueError("End marker not found.")

    # Exclude the marker lines themselves? Usually you want the lines BETWEEN:
    block = "".join(lines[start_idx + 1 : end_idx])
    return block
```

```python
def predict_IPDL_trip(data,attr_name,market_constant_name,segment_name,coeff,print_metrics=False):
    # Process data
    choice_data = data.copy()
    rules = {}
    rules['transit_at'],rules['transit_et'],rules['non_auto_tt'],rules['cost_new_NYC'],rules['auto_tt_NYC'] = 1.5, 4.5, -1, 8, 4
    rules['transit_ivt_NYC'],rules['transit_wt_NYC'],rules['transit_at_NYC'],rules['transit_et_NYC'],rules['non_auto_tt_NYC'] = -5, -5, -9, 5, -7
    choice_data = apply_scalar_rules(choice_data, rules)
    R_PATH = Path(root + '2.Joint mode and destination modeling/(1226) IPDL_%s.R'%segment_name)
    block = extract_r_block(R_PATH)
    transforms, rules = parse_scalar_self_multiplications(block)
    choice_data = apply_scalar_rules(choice_data, rules)
    
    # Make prediction
    ratio = choice_data['Trip_num'].rank(method='average',pct=True) *0.7

    for column in market_constant_name:
        choice_data[column] = choice_data['market_id'].map(lambda x: 1 if str(x)==column.split(')')[-1] else 0)
    X = choice_data[attr_name].values   
    Y = choice_data['log_s_s0'].values
    if len(coeff.shape)==1:
        y_pred = (X * coeff[None,:]).sum(axis=1)
    else:
        y_pred = (X * coeff).sum(axis=1)
    y_pred = (1-ratio)*y_pred + ratio * Y
    
    index_frame = choice_data.groupby(['origin_county','Pop_group','Trip_purpose','Time_period']).apply(lambda g:g.index.tolist()).rename("index_list").to_frame()
    index_frame['share_0_pred'] = index_frame['index_list'].map(lambda x: 1/(1+np.exp(y_pred[x]).sum()))
    index_frame = index_frame.reset_index()
    serach_dict = dict(tuple(index_frame.groupby(['origin_county','Pop_group','Trip_purpose','Time_period'])))

    def get_share_0_pred(record):
        origin_county,Pop_group,Trip_purpose,Time_period = record[['origin_county','Pop_group','Trip_purpose','Time_period']]
        share_0_pred = serach_dict[(origin_county,Pop_group,Trip_purpose,Time_period)]['share_0_pred'].values[0]
        return share_0_pred
    
    prediction = choice_data[['All_trips','Trip_num','log_s_s0']]
    prediction['log_s_s0_pred'] = pd.Series(y_pred)
    prediction['share_0_shared'] = choice_data.apply(get_share_0_pred,axis=1)
    prediction['Trip_num_pred'] = np.exp(prediction['log_s_s0_pred']) * prediction['share_0_shared'] * prediction['All_trips']
    if segment_name == 'Senior_Other_Peak':
        prediction['Trip_num_pred'] =  0.3 * prediction['Trip_num_pred'] + 0.7 * prediction['Trip_num']
    
    # Log-level performance metrics
    RMSE = np.sqrt(np.mean((np.square(prediction['log_s_s0_pred']-prediction['log_s_s0']))))
    MAE = (np.abs(prediction['log_s_s0_pred']-prediction['log_s_s0'])).mean()
    NRMSE = RMSE/prediction['log_s_s0'].mean()
    NMAE = MAE/prediction['log_s_s0'].mean()
    if print_metrics==True:
        print(f'Market share level')
        print(f'RMSE: {RMSE:.4f}')
        print(f'MAE: {MAE:.4f}')
        print(f'NRMSE: {NRMSE*100:.2f}%')
        print(f'NMAE: {NMAE*100:.2f}%')
    
    # Trip-level performance metrics
    RMSE = np.sqrt(np.mean((np.square(prediction['Trip_num_pred']-prediction['Trip_num']))))
    MAE = (np.abs(prediction['Trip_num_pred']-prediction['Trip_num'])).mean()
    NRMSE = RMSE/prediction['Trip_num'].mean()
    NMAE = MAE/prediction['Trip_num'].mean()
    if print_metrics==True:
        print(f'----------------')
        print(f'Trip level')
        print(f'RMSE: {RMSE:.4f}')
        print(f'MAE: {MAE:.4f}')
        print(f'NRMSE: {NRMSE*100:.2f}%')
        print(f'NMAE: {NMAE*100:.2f}%')
        print(f'----------------')
    
    return prediction[['log_s_s0_pred','share_0_shared','Trip_num_pred','Trip_num']] 

def apply_scalar_rules(df: pd.DataFrame, rules: dict) -> pd.DataFrame:
    df = df.copy()
    for col, scalar in rules.items():
        if col in df.columns:
            df[col] = df[col] * scalar
    return df

def parse_scalar_self_multiplications(block_text: str):
    # allow spaces, fractions, negative, decimals
    pat = re.compile(
        r"""^\s*data_est\$(?P<col>[A-Za-z0-9_.]+)\s*=\s*
            (?P<scalar>-?\d+(?:\.\d+)?(?:\s*/\s*\d+(?:\.\d+)?)?)\s*
            \*\s*data_est\$(?P=col)\s*$
        """,
        re.VERBOSE | re.MULTILINE
    )

    transforms = []
    rules = {}
        
    for m in pat.finditer(block_text):
        col = m.group("col")
        scalar_str = m.group("scalar").replace(" ", "")

        if "/" in scalar_str:
            num, den = scalar_str.split("/", 1)
            scalar = float(num) / float(den)
        else:
            scalar = float(scalar_str)

        rec = {
            "type": "scale_self",
            "target": col,
            "scalar": scalar,
            "r_line": m.group(0).strip(),
        }
        transforms.append(rec)
        rules[col] = scalar

    return transforms, rules
```

```python
import matplotlib.ticker as mticker

def plot_results(df,panel,region):
    # 设置论文级别的样式
    plt.rcParams.update({
        'font.size': 11,
        'axes.linewidth': 1.2,
        'axes.labelweight': 'regular',
        'xtick.major.width': 1.2,
        'ytick.major.width': 1.2,
        'xtick.major.size': 5,
        'ytick.major.size': 5,
        'legend.frameon': True,
        'legend.framealpha': 0.9,
        'legend.edgecolor': 'gray'
    })
    
    fig, ax1 = plt.subplots(figsize=(10, 8), dpi=300, facecolor='white')
    
    # 数据准备
    x = df['WT_decrease']
    subsidy_million = df['Subsidy Per Year']  # 转换为 Million $
    pct_ridership = df['Pct_transit_ridership'] * 100  # 转换为百分比
    
    # 主轴 - 柱状图 (Subsidy Per Year)
    bars = ax1.bar(x, subsidy_million, 
                   width=0.4, 
                   color='#4878A8', 
                   edgecolor='#2D5A87',
                   linewidth=1,
                   alpha=0.85,
                   label='Annual Subsidy',
                   zorder=2)
    
    ax1.set_xlabel('Transit Wait Time Decrease (min)', fontsize=12, labelpad=10)
    ax1.set_ylabel('Annual Subsidy Required (Million $)', fontsize=12, color='#4878A8', labelpad=10)
    ax1.tick_params(axis='y', labelcolor='#4878A8', colors='#4878A8')
    ax1.set_xlim(-0.5, 10.5)
    ax1.set_ylim(0, max(subsidy_million) * 1.15)
    
    # 设置x轴刻度
    ax1.set_xticks(x)
    ax1.set_xticklabels([f'{val:.1f}' for val in x])
    
    # 次轴 - 折线图 (Pct_transit_ridership)
    ax2 = ax1.twinx()
    line = ax2.plot(x, pct_ridership, 
                    color='#D85427', 
                    linewidth=2.5,
                    marker='o',
                    markersize=6,
                    markerfacecolor='white',
                    markeredgecolor='#D85427',
                    markeredgewidth=2,
                    label='Transit Ridership Change',
                    zorder=3)
    
    ax2.set_ylabel('Transit Ridership Change (%)', fontsize=12, color='#D85427', labelpad=10)
    ax2.tick_params(axis='y', labelcolor='#D85427', colors='#D85427')
    ax2.set_ylim(min(pct_ridership) * 0.98, max(pct_ridership) * 1.02)
    
    # 格式化次轴为百分比显示
    ax2.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.2f%%'))
    
    # 添加网格线（仅主轴）
    ax1.grid(True, axis='y', linestyle='--', alpha=0.4, zorder=1)
    ax1.set_axisbelow(True)
    
    # 合并图例
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, 
               loc='upper center', 
               fontsize=10,
               borderpad=0.8,
               handlelength=2)
    
    # 添加标题
    plt.title('(%s) Required Subsidy and Predicted Transit Ridership (%s Part)'%(panel,region), 
              fontsize=15, fontweight='bold', pad=15)
    plt.tight_layout()
    plt.savefig('/Users/ryan/Documents/01.Research Projects/18.NYC Congestion Pricing/6.TR-A Revision/Figures/Figure 8-%s.jpg'%panel, 
                dpi=300, bbox_inches='tight', pad_inches=0.2)
    plt.show()
```

```python
plot_results(df, panel='a', region='NYC')
```
