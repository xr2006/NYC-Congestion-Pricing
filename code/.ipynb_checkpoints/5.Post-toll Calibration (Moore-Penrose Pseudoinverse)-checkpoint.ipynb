### Required functions

```python
def compute_jacobian(
    objective_components: Callable,
    x0: np.ndarray,
    epsilon: float = 1e-5
) -> np.ndarray:
    """
    Compute the Jacobian matrix of the objective components with respect to
    decision variables using finite differences.
    
    The Jacobian J[i,j] = ∂f_i / ∂x_j tells us how each output (target metric)
    changes with respect to each input (decision variable).
    
    Parameters
    ----------
    objective_components : Callable
        Function that takes decision variables and returns a vector of 
        the individual metric values (not squared errors).
    x0 : np.ndarray
        Point around which to linearize (typically zeros or current estimate).
    epsilon : float
        Step size for finite differences.
        
    Returns
    -------
    J : np.ndarray
        Jacobian matrix of shape (n_outputs, n_inputs)
    """
    n_vars = len(x0)
    f0 = objective_components(x0)
    n_outputs = len(f0)
    
    J = np.zeros((n_outputs, n_vars))
    
    for j in range(n_vars):
        x_plus = x0.copy()
        x_plus[j] += epsilon
        x_minus = x0.copy()
        x_minus[j] -= epsilon
        
        f_plus = objective_components(x_plus)
        f_minus = objective_components(x_minus)
        
        # Central difference approximation
        J[:, j] = (f_plus - f_minus) / (2 * epsilon)
    
    return J
```

```python
def solve_with_pseudoinverse(
    J: np.ndarray,
    targets: np.ndarray,
    current_values: np.ndarray,
    constraint_matrix: np.ndarray = None,
    constraint_rhs: np.ndarray = None
) -> Tuple[np.ndarray, Dict]:
    """
    Solve for decision variables using Moore-Penrose pseudoinverse.
    
    We want to find x such that J @ x ≈ (targets - current_values)
    
    For underdetermined systems, the pseudoinverse gives the minimum-norm solution.
    
    Parameters
    ----------
    J : np.ndarray
        Jacobian matrix (n_targets × n_variables)
    targets : np.ndarray
        Target values for each metric
    current_values : np.ndarray
        Current values of metrics at x=0
    constraint_matrix : np.ndarray, optional
        Matrix C for equality constraints C @ x = d
    constraint_rhs : np.ndarray, optional
        Right-hand side d for equality constraints
        
    Returns
    -------
    x : np.ndarray
        Minimum-norm solution
    info : dict
        Additional information about the solution
    """
    # The linearized system is: J @ x = b, where b = targets - current_values
    b = targets - current_values
    
    if constraint_matrix is not None:
        # Solve constrained minimum-norm problem using KKT conditions
        # minimize ||x||^2 subject to J @ x = b and C @ x = d
        # 
        # This is equivalent to finding the minimum-norm solution to:
        # [J]     [b]
        # [C] x = [d]
        
        A_augmented = np.vstack([J, constraint_matrix])
        b_augmented = np.hstack([b, constraint_rhs])
        
        x = np.linalg.pinv(A_augmented) @ b_augmented
    else:
        # Standard pseudoinverse solution
        x = np.linalg.pinv(J) @ b
    
    # Compute solution quality metrics
    residual = J @ x - b
    
    info = {
        'jacobian': J,
        'target_residuals': residual,
        'residual_norm': np.linalg.norm(residual),
        'solution_norm': np.linalg.norm(x),
        'jacobian_rank': np.linalg.matrix_rank(J),
        'singular_values': np.linalg.svd(J, compute_uv=False),
        'predicted_values': current_values + J @ x
    }
    
    return x, info
```

```python
def iterative_pseudoinverse_solve(
    objective_components: Callable,
    targets: np.ndarray,
    x0: np.ndarray = None,
    constraint_matrix: np.ndarray = None,
    constraint_rhs: np.ndarray = None,
    max_iters: int = 20,
    tol: float = 1e-8,
    verbose: bool = True
) -> Tuple[np.ndarray, Dict]:
    """
    Iteratively solve using pseudoinverse with Jacobian updates.
    
    Since the relationship between decision variables and outcomes may be
    nonlinear, we iterate: linearize → solve → update → repeat.
    
    This is essentially Gauss-Newton for minimum-norm solutions.
    
    Parameters
    ----------
    objective_components : Callable
        Function returning vector of metric values
    targets : np.ndarray
        Target values
    x0 : np.ndarray, optional
        Initial guess (default: zeros)
    constraint_matrix : np.ndarray, optional
        Equality constraint matrix C (for C @ x = d)
    constraint_rhs : np.ndarray, optional  
        Equality constraint RHS d
    max_iters : int
        Maximum iterations
    tol : float
        Convergence tolerance on residual norm
    verbose : bool
        Print iteration progress
        
    Returns
    -------
    x : np.ndarray
        Solution
    history : dict
        Convergence history and diagnostics
    """
    n_vars = len(x0) if x0 is not None else 6  # default for your problem
    if x0 is None:
        x0 = np.zeros(n_vars)
    
    x = x0.copy()
    history = {
        'x': [x.copy()],
        'residuals': [],
        'solution_norms': []
    }
    
    for iteration in range(max_iters):
        # Evaluate current metrics
        current_values = objective_components(x)
        residual = targets - current_values
        residual_norm = np.linalg.norm(residual)
        
        history['residuals'].append(residual_norm)
        history['solution_norms'].append(np.linalg.norm(x))
        
        if verbose:
            print(f"Iteration {iteration}: residual norm = {residual_norm:.2e}, "
                  f"solution norm = {np.linalg.norm(x):.4f}")
        
        if residual_norm < tol:
            if verbose:
                print(f"Converged at iteration {iteration}")
            break
        
        # Compute Jacobian at current point
        J = compute_jacobian(objective_components, x)
        
        # Solve for update delta_x
        if constraint_matrix is not None:
            A_aug = np.vstack([J, constraint_matrix])
            b_aug = np.hstack([residual, constraint_rhs - constraint_matrix @ x])
            delta_x = np.linalg.pinv(A_aug) @ b_aug
        else:
            delta_x = np.linalg.pinv(J) @ residual
        
        # Update (with optional damping for stability)
        x = x + delta_x
        history['x'].append(x.copy())
    
    # Final evaluation
    final_values = objective_components(x)
    history['final_values'] = final_values
    history['final_residual'] = targets - final_values
    history['jacobian_final'] = compute_jacobian(objective_components, x)
    
    return x, history
```

```python
def objective_components_real(decision_variables):
    """
    Compute the three target metrics given decision variables.
    
    This wraps your existing prediction_with_coeff_adj function.
    
    Parameters
    ----------
    decision_variables : array-like
        [constant_driving, constant_transit, constant_ondemand, 
         constant_biking, constant_walking, asc_crz]
    
    Returns
    -------
    metrics : np.ndarray
        [pct_change_vehicle_entry, pct_change_ondemand, pct_change_transit]
    """
    constant_driving, constant_transit, constant_ondemand, constant_biking, constant_walking, asc_crz = decision_variables
    
    adj = {
        'constant_driving': constant_driving,
        'constant_transit': constant_transit,
        'constant_ondemand': constant_ondemand,
        'constant_biking': constant_biking,
        'constant_walking': constant_walking,
        'County_36061-1': asc_crz
    }
    
    # Make predictions for 2023 and 2025
    trip_pred_25 = prediction_with_coeff_adj(choice_data_25, Segment_lst, adj=adj)
    trip_pred_23 = prediction_with_coeff_adj(choice_data_23, Segment_lst)
    
    # Conditions (same as your original code)
    condition1 = choice_data['mode'].isin(['driving', 'ondemand', 'carpool'])
    condition2 = choice_data['destination_county'] == '36061-1'
    condition3 = choice_data['origin_county'] == '36061-1'
    
    # Compute percentage changes
    CRZ_25 = trip_pred_25['Trip_num_pred'][condition1 & condition2].sum()
    CRZ_23 = trip_pred_23['Trip_num_pred'][condition1 & condition2].sum()
    pct_change_vehicle_entry = (CRZ_25 - CRZ_23) / CRZ_23
    
    Ondemand_25 = trip_pred_25['Trip_num_pred'][(choice_data['mode']=='ondemand') & (condition2|condition3)].sum()
    Ondemand_23 = trip_pred_23['Trip_num_pred'][(choice_data['mode']=='ondemand') & (condition2|condition3)].sum()
    pct_change_ondemand = (Ondemand_25 - Ondemand_23) / Ondemand_23
    
    Transit_25 = trip_pred_25['Trip_num_pred'][(choice_data['mode']=='transit') & (condition2|condition3)].sum()
    Transit_23 = trip_pred_23['Trip_num_pred'][(choice_data['mode']=='transit') & (condition2|condition3)].sum()
    pct_change_transit = (Transit_25 - Transit_23) / Transit_23
    
    return np.array([pct_change_vehicle_entry, pct_change_ondemand, pct_change_transit])
```

```python
def prediction_with_coeff_adj(data,Segment_lst,adj={'constant_driving':0,'constant_transit':0,'constant_ondemand':0,'constant_biking':0,'constant_walking':0,'County_36061-1':0}):
    # initialize columns once (optional but recommended)
    choice_data = data.copy()
    pred_cols = ['log_s_s0_pred', 'share_0_shared', 'Trip_num_pred']
    # For loop segments
    for segment_name in Segment_lst:
        pop_group, trip_purpose, time_period = segment_name.split('_')
        # ---- load coefficients ----
        file_path = (
            root
            + f'2.Joint mode and destination modeling/Model Results 1226/'
              f'IPDL_{segment_name}.xlsx'
        )
        df = pd.read_excel(file_path, sheet_name="Coefficients")
        attr_name = df["term"].tolist()
        attr_name = [s.replace('log_s_s.county.', 'log_s_s(county)') for s in attr_name]
        attr_name = [s.replace('log_s_s.mode.', 'log_s_s(mode)') for s in attr_name]
        attr_name = [s.replace('County_36061.0', 'County_36061-0') for s in attr_name]
        attr_name = [s.replace('County_36061.1', 'County_36061-1') for s in attr_name]
        coeff = df['estimate'].values
        market_constant_name = [x for x in attr_name if x.startswith('factor')]
        mask = (
            (choice_data['Pop_group'] == pop_group) &
            (choice_data['Trip_purpose'] == trip_purpose) &
            (choice_data['Time_period'] == time_period)
        )
        piece = choice_data.loc[mask]
        
        # ---- apply adjustment ---
        coeff_full = np.tile(coeff, (len(piece), 1))
        coeff_adj = coeff_full.copy()
        # Mode constant adjustment is only for trips from/to CRZ
        condition = (piece['destination_county']=='36061-1')|(piece['origin_county']=='36061-1')
        for attr in list(adj.keys())[:-1]:
            coeff_adj[condition,attr_name.index(attr)] += adj[attr]
        # County constant adjustment is for all trips
        coeff_adj[:,attr_name.index('County_36061-1')] += adj['County_36061-1']

        # ---- predict ----
        preds = predict_IPDL_trip(
            piece,
            attr_name,
            market_constant_name,
            segment_name,
            coeff_adj,
            print_metrics = False
        )
    
        # preds must be aligned row-wise with `piece`
        # ---- assign back safely ----
        choice_data.loc[mask, pred_cols] = preds.iloc[:,:-1].values
    
    return choice_data
```
