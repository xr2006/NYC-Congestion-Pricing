### MTA Bridge

```python
data = pd.read_csv('/Users/ryan/Documents/01.Research Projects/18.NYC Congestion Pricing/3.Model Calibration/(1226) Post-toll Calibration/MTA_Bridges_and_Tunnels_Hourly_Crossings__Beginning_2019_20260122.csv')
```

```python
data['Month'] = data['Transit Timestamp'].map(lambda x: x.split(' ')[0].split('/')[0])
data['Year'] = data['Transit Timestamp'].map(lambda x: x.split(' ')[0].split('/')[-1])
data = data[(data['Year'].isin(['2022','2023','2024','2025'])) & data['Month'].isin(['04','05','06'])]
#data = data[(data['Facility ID'].isin([27,28]))&(data['Vehicle Class Category']=='Car')]
# data = data[data['Direction'].isin(['Northbound to Manhattan','Westbound to Manhattan'])]
```
